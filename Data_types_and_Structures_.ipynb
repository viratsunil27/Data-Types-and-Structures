2. Explain the difference between mutable and immutable data types with examples

Mutable:
Objects whose values can be changed without changing their identity.

Example: List
,Dictionary
,Set.

Immutable:
Objects whose content cannot be changed after creation. Any "change" results in a new object.

Example: Int, Float, string, tuple.
